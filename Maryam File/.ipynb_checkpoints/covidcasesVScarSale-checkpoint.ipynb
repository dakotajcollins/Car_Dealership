```python
 import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = "../CSV/covid_us_county.csv"

covid_df = pd.read_csv(file)
covid_df.head()
```

```python
case_total = covid_df.groupby(["state"]).sum()["cases"]
mean_state=covid_df.groupby(["state"]).mean()["cases"]

covid_by_state=pd.DataFrame({"Cases":case_total,
                             "Mean" :mean_state})

                             
covid_by_state                                  
```

```python
covid_by_state.plot(kind="bar",)
plt.xlabel('State')
plt.ylabel('Number of Positive covid tests in 2020')
plt.figure (figsize=(20,40))
plt.show()
```

```python
Ten_top=pd.DataFrame(covid_by_state.nlargest(10, ['Cases']))
Ten_top
```

```python
Ten_top.plot(kind="bar",)
plt.xlabel('State')
plt.ylabel('Number of Positive covid tests in 2020')
plt.figure (figsize=(20,40))
plt.show()
```
